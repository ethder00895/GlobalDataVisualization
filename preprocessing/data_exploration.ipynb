```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace df with your DataFrame
missing = df.isnull().sum()
present = df.notnull().sum()

plot_df = pd.DataFrame({
    'Missing': missing,
    'Present': present
})

fig, ax = plt.subplots(figsize=(14, 6))

percent_df = (plot_df.div(plot_df.sum(axis=1), axis=0) * 100)

# Plot stacked bars
bars = percent_df.plot(
    kind='bar',
    stacked=True,
    ax=ax,
    color=['crimson', 'seagreen'],
    edgecolor='black'
)

# Horizontal dashed gridlines
ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
ax.set_axisbelow(True)

# Add labels inside each bar segment
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%d',
        label_type='center',
        color='white',
        fontsize=9
    )

# Move legend outside the plot
ax.legend(
    title="Value Type",
    bbox_to_anchor=(1.02, 1),
    loc='upper left',
    borderaxespad=0
)

# Titles and labels
plt.title("Missing vs Present Values per Column")
plt.xlabel("Columns")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## Step 1 — Calculate Missingness Per Country

```python
core_cols = [
    'Access to electricity (% of population)',
    'Access to clean fuels for cooking',
    'Renewable energy share in the total final energy consumption (%)',
    'Electricity from fossil fuels (TWh)',
    'Electricity from nuclear (TWh)',
    'Electricity from renewables (TWh)',
    'Low-carbon electricity (% electricity)',
    'Primary energy consumption per capita (kWh/person)',
    'Energy intensity level of primary energy (MJ/$2017 PPP GDP)',
    'Value_co2_emissions_kt_by_country',
    'gdp_growth',
    'gdp_per_capita'
]

country_missing = (
    df.groupby('Entity')[core_cols]
      .apply(lambda x: x.isna().mean() * 100)
      .round(2)
)

country_missing
```

```python
# missing values in percentage
country_missing.to_excel('./data/missing.xlsx')
```

## Step 2 — Calculate Overall Completeness Score

```python
country_missing['overall_missing_pct'] = (
    country_missing.mean(axis=1)
)

country_missing['overall_completeness_pct'] = (
    100 - country_missing['overall_missing_pct']
)

```

## Step 3 — Identify Weak Countries

```python
weak_countries = country_missing[
    country_missing['overall_completeness_pct'] < 70
]

weak_countries
```

## Step 4 — Review Before Dropping

```python
print(weak_countries.sort_values(
    'overall_completeness_pct'
))
```

```python
country_missing = (
    df.groupby('Entity')[core_cols]
      .apply(lambda x: x.isna().mean() * 100)
)

country_missing['overall_completeness_pct'] = (
    100 - country_missing.mean(axis=1)
)

country_missing.sort_values(
    'overall_completeness_pct'
)
```

## Overall Completeness of Data in Percentage

```python
country_pct = (
    country_missing
    .groupby('Entity')['overall_completeness_pct']
    .mean()
    .sort_values(ascending=False)
)

# print(country_pct)

for country, pct in country_pct.items():
    print(f"{country}: {pct:.2f}%")
```
